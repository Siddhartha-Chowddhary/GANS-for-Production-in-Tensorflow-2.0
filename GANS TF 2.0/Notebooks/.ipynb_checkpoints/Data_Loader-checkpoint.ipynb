```python
from __future__ import absolute_import, division, print_function, unicode_literals



import matplotlib.pyplot as plt
import numpy as np
import os


import time

from IPython import display
```

```python
dir_data      = "Path/To/Data"
Ntrain        = 6161 
# Ntest         = 100
nm_imgs       = np.sort(os.listdir(dir_data))
## name of the jpg files for training set
nm_imgs_train = nm_imgs[:Ntrain]
## name of the jpg files for the testing data
img_shape     = (64, 64, 3)
```

```python
#Generating Numpy Data

from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True


def Data_Generator(nm_imgs_train):
    Training_data = []
    for i, myid in enumerate(nm_imgs_train):
        images = load_img(dir_data + "/" + myid,
                         target_size=img_shape[:2])
        images = img_to_array(images)/255.0
        Training_data.append(images)
    Training_data = np.array(Training_data)
    return(Training_data)

Training_data = Data_Generator(nm_imgs_train)
print("Training_data.shape = {}".format(Training_data.shape))
```

```python
# Resized Data Visualizer
fig = plt.figure(figsize=(30,10))
nplot = 10
for count in range(1,nplot):
    ax = fig.add_subplot(1,nplot,count)
    ax.imshow(Training_data[count])
plt.show()
```

```python
# Saving the Resized image data
saving_dir = "Path/To/Saving_directory"
# np.save('Training_Data.npy', Training_data)
np.save(os.path.join(saving_dir, 'Training_Data.npy'), Training_data)
```
